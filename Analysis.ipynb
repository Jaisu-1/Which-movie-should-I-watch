### Normalize both imdb rating and metascore

```python
data['n_imdb'] = data['Imdb'] * 10
```

```python
data['n_imdb'].head(5)
```

```python
data.to_csv('movie_ratings.csv')
```

```python
%matplotlib inline
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 3, figsize = (30,10))
ax1, ax2, ax3 = fig.axes

ax1.hist(data['Imdb'], bins = 30, range = (0,10)) # bin range = 1
ax1.set_title('IMDB rating')

ax2.hist(data['Metascore'], bins = 30, range = (0,100)) # bin range = 10
ax2.set_title('Metascore')

ax3.hist(data['n_imdb'], bins=30, range = (0,100), histtype = 'step')
ax3.hist(data['Metascore'], bins=30, range = (0,100), histtype  = 'step')
ax3.legend(loc = 'upper left')
ax3.set_title('The Two Normalized Distributions')


for ax in fig.axes:
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

ax1, 
```
